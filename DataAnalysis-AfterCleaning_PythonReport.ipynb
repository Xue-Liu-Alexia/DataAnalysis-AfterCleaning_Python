## a) Read Energy Data

Read the energy data from the file Energy Indicators.xls, and put it into a Data Frame with the variable name of energy.
- This is an Excel file, use 'pd.read_excel()' to read this data and use `'skiprows'` and `'skipfooter'` to exclude the header and footer information, also there have some missing data, reflect them as `NaN` values.

```python
import pandas as pd
```

```python
energy = pd.read_excel('D:\DataScience\Python\Project\EnergyIndicators.xls', \
                       header= None, na_values='...', skiprows=range(18), \
                       skipfooter=(len(pd.read_excel('D:\DataScience\Python\Project\EnergyIndicators.xls'))-244))
energy
```

```python
# delete first two columns 
energy = energy.iloc[:, 2:]
energy
```

## c) Rename Country Names

Rename 4 countries, there are several countries with parenthesis in their name, remove the parenthesis.
<br>1. First check the original country name and find that some country names have a `number suffix` after it, use `str.replace()` to remove the suffix first.
<br>2. Then change the country name and use `str.replace()` to remove parenthesis.
<br>3. Finally check the country name and compare it with the original country name.

```python
# original country name
energy['Country'].unique()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# remove all the numbers from the 'Country' column
energy['Country'] = energy['Country'].str.replace('\d+', '')
energy['Country'].unique()
```

```python
# rename countries, remove the parenthesis 
country_rename = {
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"}
energy['Country'] = energy['Country'].replace(country_rename)
energy['Country'] = energy['Country'].str.replace(r" \(.*\)", "")
energy['Country'].unique()
```

## <font color = green> Q10: Group the Countries by Continent and display some descriptive statistics measures. </font>

First, create a data frame with an index named 'Continent' according to the division of continents.
<br> Then use `groupby()`and `map()` to group the Countries by Continent.
<br> Use `size() sum() mean() std()` to display the sample size, sum, mean, and std deviation for the estimated population of each continent.

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
```

```python
Continent = pd.DataFrame(index=['Asia', 'Australia', 'Europe', 'North America', 'South America'], \
                         columns=['size', 'sum', 'mean', 'std'])
Continent.index.name = 'Continent'
```

```python
continent_group = df.groupby(df.index.map(ContinentDict))['Population Estimate']
Continent['size'] = continent_group.size()
Continent['sum'] = continent_group.sum()
Continent['mean'] = continent_group.mean()
Continent['std'] = continent_group.std()
Continent
```

## <font color = green> Q11: Cut % Renewable into 5 bins. How many countries are in each of these groups? </font>

Use `pd.cut(bins=5)` to cut % Renewable into 5 bins,  use `groupby().size()` to group these new % Renewable bins.
<br>Return a Series, but not include groups with no countries.

```python
new_df = df[['% Renewable']]
new_df['Renewable Bin'] = pd.cut(new_df['% Renewable'], bins=5)
MultiIndex_Continent = new_df.groupby([df.index.map(ContinentDict), \
                                       'Renewable Bin']).size()
# not include groups with no countries
MultiIndex_Continent = MultiIndex_Continent[MultiIndex_Continent > 0] 
MultiIndex_Continent
```
